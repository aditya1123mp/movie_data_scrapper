```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.action_chains import ActionChains
import time
import pandas as pd

# Open browser
driver = webdriver.Chrome()

# Maximize window
driver.maximize_window()

# Navigate to IMDb website
driver.get('http://www.imdb.com/chart/top')

# Delay for 3 seconds to let the page load
time.sleep(3)

# Refresh the current web page
driver.refresh()

# Wait for the elements to be visible (up to 10 seconds)
wait = WebDriverWait(driver, 10)
wait.until(EC.presence_of_all_elements_located((By.XPATH, '//button[@class="ipc-icon-button cli-info-icon ipc-icon-button--base ipc-icon-button--onAccent2"]')))

# Find elements by XPath
elements = driver.find_elements(By.XPATH, '//button[@class="ipc-icon-button cli-info-icon ipc-icon-button--base ipc-icon-button--onAccent2"]')

# Get the count of elements
count = len(elements)
print(f'Number of elements found: {count}')

# Initialize lists to store the extracted data
movie_titles = []
release_years = []
runtimes = []
ratings = []
film_genres = []
IMDB_ratings = []
movie_stories = []
directors = []
stars = []

# Loop over all the found elements
for i in range(count):
    try:
        # Re-find elements in each iteration to avoid stale element exception
        elements = driver.find_elements(By.XPATH, '//button[@class="ipc-icon-button cli-info-icon ipc-icon-button--base ipc-icon-button--onAccent2"]')

        # Ensure the element is within the viewport
        driver.execute_script("arguments[0].scrollIntoView(true);", elements[i])

        # Wait until the element is clickable
        WebDriverWait(driver, 10).until(EC.element_to_be_clickable(elements[i]))

        # Use JavaScript to click the element
        driver.execute_script("arguments[0].click();", elements[i])

        # Wait for the modal to appear
        time.sleep(3)  # Adjust the time if needed

        # Extract movie data
        movie_title = driver.find_element(By.XPATH, "//*[@class='ipc-title__text prompt-title-text']").text
        release_year = driver.find_element(By.XPATH, "//ul[@data-testid='btp_ml']/li[1]").text
        runtime = driver.find_element(By.XPATH, "//ul[@data-testid='btp_ml']/li[2]").text
        rating = driver.find_element(By.XPATH, "//ul[@data-testid='btp_ml']/li[3]").text
        genre = driver.find_element(By.XPATH, "//ul[@data-testid='btp_gl']/li[1]").text
        imdb_rating = driver.find_element(By.XPATH, "//span[@class='ipc-rating-star ipc-rating-star--baseAlt ipc-rating-star--imdb btp_rt_ds']/span[1]").text
        movie_story = driver.find_element(By.XPATH, "//div[@class='sc-65f72df1-2 geapts']").text
        director = driver.find_element(By.XPATH, "//div[@data-testid='p_ct_dr']//ul/li/a").text
        star_elements = driver.find_elements(By.XPATH, "//div[@data-testid='p_ct_cst']//ul/li/a")

        # Get the names of the stars
        star_names = ', '.join([star.text for star in star_elements])

        # Append the extracted data to the lists
        movie_titles.append(movie_title)
        release_years.append(release_year)
        runtimes.append(runtime)
        ratings.append(rating)
        film_genres.append(genre)
        IMDB_ratings.append(imdb_rating)
        movie_stories.append(movie_story)
        directors.append(director)
        stars.append(star_names)

        # Close the modal
        close_button = driver.find_element(By.XPATH, "//button[@title='Close Prompt' and @aria-label='Close Prompt']")
        driver.execute_script("arguments[0].click();", close_button)
        time.sleep(2)  # Small delay before continuing to the next iteration

    except Exception as e:
        print(f"Error during iteration {i + 1}: {e}")
        continue

# Close the browser
driver.quit()

# Create a DataFrame to store the data
df = pd.DataFrame({
    'Movie Title': movie_titles,
    'Release Year': release_years,
    'Runtime': runtimes,
    'Rating': ratings,
    'Film Genres': film_genres,
    'IMDB Rating': IMDB_ratings,
    'Movie Story': movie_stories,
    'Director': directors,
    'Stars': stars
})

# Save the DataFrame to the Excel file
output_file_path = r'C:\Users\DELL\OneDrive\Desktop\mail\IMDB_movie_data.xlsx'
df.to_excel(output_file_path, index=False)

print("Data saved to IMDB_movie_data.xlsx")
```
